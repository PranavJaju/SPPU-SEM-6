```python
file = pd.read_csv("./nba.csv")
file
```

```python
file['Age'].value_counts()
```

```python
mode = file['College'].mode()[0]
file['College'].fillna(mode,inplace=True)
file['Salary'].fillna(file['Salary'].mean(),inplace=True)
```

```python
file.isnull().sum()
```

```python
file.dropna(inplace=True)
file.isnull().sum()
```

```python
file.reset_index(drop=True,inplace=True)
```

```python
bins = [19,24,29,35,40]
labels = ['19-23','24-28','29-34','35-40']
file['AgeGroup'] = pd.cut(file['Age'],bins = bins,labels = labels, right = False)
```

```python
file['AgeGroup'].value_counts()
```

```python
file.groupby('AgeGroup',observed=False).describe().transpose()
```

```python
List_of_cat  = (list(file['AgeGroup'].value_counts().index))
List_of_cat
```
